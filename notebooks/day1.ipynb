```python
import torch
from transformers import CLIPModel, CLIPProcessor

MODEL_NAME = "patrickjohncyh/fashion-clip"

model = CLIPModel.from_pretrained(MODEL_NAME)
processor = CLIPProcessor.from_pretrained(MODEL_NAME)
model.eval()

print("모델 로드 완료")
```

```python
from PIL import Image

image = Image.open("../data/raw/test.png").convert("RGB")

inputs = processor(images=image, return_tensors="pt")

import torch
with torch.no_grad():
    out = model.get_image_features(**inputs)

features = out if torch.is_tensor(out) else out.pooler_output

print("반환 타입:", type(out).__name__)
print("벡터 shape:", features.shape)
print("앞 5개 값:", features[0][:5])


```

```python
import torch.nn.functional as F

normalized = F.normalize(features, p=2, dim=1)
print("정규화 후 길이: ", normalized.norm().item())
```
